```python
cicero_texts = []
counter = 0
for text in cicero:
    for item in text:
        counter += 1
        print(counter)
        try:
            cicero_texts.append(list(item.values()))
        except AttributeError:
            cicero_texts.append(item)
```

```python
cicero_strings = []
for item in cicero_texts:
    if type(item) == list:
        if dict in [type(it) for it in item]:
            continue
        cicero_strings.append(''.join(item))
    elif type(item) == str:
        cicero_strings.append(item)
    else:
        raise TypeError
```

```python
long_cicero = ''.join(cicero_strings)
```

```python
cicero_sentences1 = long_cicero.split('.')
```

```python
last_cicero, last_caesar, last_ovid = [],[],[]
for c in final_cicero:
    last_cicero.append(lemmatizer.lemmatize(c))
for c in final_caesar:
    last_caesar.append(lemmatizer.lemmatize(c))
for o in ultimate_ovid:
    last_ovid.append(lemmatizer.lemmatize(o))
```

```python
j = JVReplacer()
cicero_stops, caesar_stops, ovid_stops = [],[],[]
for lis in last_cicero:
    inter=[]
    for word in lis:
        if word[0] not in STOPS_LIST:
            inter.append(j.replace(word[0]))
    cicero_stops.append(inter)
for lis in last_caesar:
    inter=[]
    for word in lis:
        if word[0] not in STOPS_LIST:
            inter.append(j.replace(word[0]))
    caesar_stops.append(inter)
for lis in last_ovid:
    inter=[]
    for word in lis:
        if word[0] not in STOPS_LIST:
            inter.append(j.replace(word[0]))
    ovid_stops.append(inter)
```

```python
import pickle
with open('cicero.pickle', 'wb') as f:
    pickle.dump(cicero_stops, f, pickle.HIGHEST_PROTOCOL)
with open('caesar.pickle', 'wb') as f:
    pickle.dump(caesar_stops, f, pickle.HIGHEST_PROTOCOL)
with open('ovid.pickle', 'wb') as f:
    pickle.dump(ovid_stops, f, pickle.HIGHEST_PROTOCOL)
```
